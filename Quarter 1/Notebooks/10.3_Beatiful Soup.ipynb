# Beautiful Soup

```python
from bs4 import BeautifulSoup
import requests
url = "https://blog.miguelgrinberg.com/post/the-flask-mega-tutorial-part-xxiii-application-programming-interfaces-apis"
r = requests.get(url)
html = r.text
```

```python
soup = BeautifulSoup(html)
```

```python
links = soup.find_all('a')
```

```python
for l in links:
    print(l.get('href'))
```
